### __Linhas de restrição__

```python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize = 25, pad = 20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

eixo.annotate('Máximo', xy=(x1,y1), fontsize=15)


eixo.axhline(max(df['temperatura']), color = 'r', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'r', linestyle='--')
```

```python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color = 'g')
eixo.set_title('Temperatura no momento', fontsize = 25, pad = 20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax() - 5000]
y2 = max(df['temperatura'] - 5)

eixo.annotate('Máximo', xy=(x1,y1), fontsize=15, 
              xytext = (x2,y2), arrowprops = dict(facecolor = 'r'))


eixo.axhline(max(df['temperatura']), color = 'r', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'r', linestyle='--')
```

```python
fig = plt.figure(figsize = (15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color = 'steelblue', alpha = 0.5)
eixo.set_title('Temperatura no momento', fontsize = 25, pad = 20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.grid(True)

x1 = df['data'][df['temperatura'].idxmax()]
y1 = max(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmax() - 5000]
y2 = max(df['temperatura'] - 5)

eixo.annotate('Máximo', xy=(x1,y1), fontsize=15, 
              xytext = (x2,y2), arrowprops = dict(facecolor = 'b'))

x1 = df['data'][df['temperatura'].idxmin()]
y1 = min(df['temperatura'])

x2 = df['data'][df['temperatura'].idxmin() - 5000]
y2 = min(df['temperatura'] + 5)

eixo.annotate('Mínimo', xy=(x1,y1), fontsize=15, 
              xytext = (x2,y2), arrowprops = dict(facecolor = 'b'))


eixo.axhline(max(df['temperatura']), color = 'b', linestyle='--')
eixo.axhline(min(df['temperatura']), color = 'b', linestyle='--')
```

```python
dfTemperatura_dia = df.groupby(df['dia_da_semana'])['temperatura'].mean()
```

```python
dfTemperatura_dia
```

```python
#ordenando o dataframe
nome_dias =['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
dfTemperatura_dia = dfTemperatura_dia[nome_dias]
```

```python
fig = plt.figure(figsize=(8,4))
eixo = fig.add_axes([0,0,1,1])
indice = range(len(dfTemperatura_dia))
cores = ['r', 'g', 'b', 'y', 'c', 'm', 'k']
eixo.bar(indice, dfTemperatura_dia, color = cores, alpha = 0.5)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=15)
eixo.set_ylabel('Temperatura média', fontsize=15)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
plt.plot()
```

```python
fig = plt.figure(figsize = (10, 7))
eixo = fig.add_axes([0,0,1,1])

explodir = [0.1,0.2,0,0,0,0,0]

eixo.pie(dfTemperatura_dia, labels = dfTemperatura_dia.index, autopct = '%.1f%%', explode=explodir, shadow = True, startangle = 90)

eixo.set_title("Temperatura por dia", fontsize = 20)
plt.plot()
```

### __Gráficos de dispersão__

```python
import numpy as np
```

```python
df = pd.read_csv('dados/iris.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize = (15, 7))
eixo = fig.add_axes([0,0,1,1])

cores = { 'Iris-setosa': 'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'], color = 'g', marker = 'o', alpha = 0.5)
eixo.set_title("Gráfico de dispersão", fontsize = 30, pad= 13)
eixo.set_xlabel('Comprimento da sépala', fontsize = 20)
eixo.set_ylabel('Largura da sépala', fontsize = 20)

eixo.axhline(df['largura_sépala'].mean(), color = 'r', linestyle='--')
eixo.axhline(df['largura_sépala'].max(), color = 'b', linestyle='--')
eixo.axhline(df['largura_sépala'].min(), color = 'b', linestyle='--')

eixo.tick_params(labelsize=15)
```

```python
fig = plt.figure(figsize = (15, 7))
eixo = fig.add_axes([0,0,1,1])

cores = { 'Iris-setosa': 'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
marcadores = { 'Iris-setosa': 'x', 'Iris-versicolor':'o', 'Iris-virginica':'v'}

#plotando para cada espécie
for especie in df['espécie'].unique():
    temp = df[df['espécie'] == especie]
    eixo.scatter(temp['comprimento_sépala'], temp['largura_sépala'], color = cores[especie]
                 ,marker = marcadores[especie])
    
    
eixo.set_title("Gráfico de dispersão", fontsize = 30, pad= 13)
eixo.set_xlabel('Comprimento da sépala', fontsize = 20)
eixo.set_ylabel('Largura da sépala', fontsize = 20)

eixo.axhline(df['largura_sépala'].mean(), color = 'r', linestyle='--')
eixo.axhline(df['largura_sépala'].max(), color = 'b', linestyle='--')
eixo.axhline(df['largura_sépala'].min(), color = 'b', linestyle='--')
eixo.legend(cores, fontsize = 20, loc = 'best')
```
