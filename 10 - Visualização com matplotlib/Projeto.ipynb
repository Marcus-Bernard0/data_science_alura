###  __1 Primeiras visualizações__ 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dados/monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df.info()
```

### Convertendo tipo da coluna data


```python
df['data'] = pd.to_datetime(df['data'])
```

```python
df.info()
```

```python
ax = plt.figure(figsize=(15, 8))
ax = plt.plot(df['data'], df['temperatura'])
ax = plt.title('Temperaturas no tempo', {'fontsize': '20'})
```

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
```

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'teal', alpha = 0.5)

eixo.set_title("Temperaturas por momento", {'fontsize': '30'})
eixo.set_ylabel("Temperatura", {'fontsize': '20'})
eixo.set_xlabel("Data", {'fontsize': '20'})
eixo.legend(['Temperatura'], loc='upper left', fontsize='17')
```

### __2 Customizando__

```python
#Lw Expessura da linha
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])
eixo.plot(df['data'], df['temperatura'], color = 'teal', alpha = 0.5, lw = 0.9)

eixo.set_title("Temperaturas por momento", {'fontsize': '30'})
eixo.set_ylabel("Temperatura", {'fontsize': '20'})
eixo.set_xlabel("Data", {'fontsize': '20'})
eixo.legend(['Temperatura'], loc='upper left', fontsize='17')
```
