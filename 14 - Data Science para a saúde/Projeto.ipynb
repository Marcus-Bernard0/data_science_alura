# 1- Introdução


```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
df.head()
```

* Limpando os dados

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df['Province/State'].unique()
```

----

```python
df.set_index('Province/State').loc['Saint Barthelemy']
```

```python
df['Province/State'].value_counts()
```

```python
confirmed_by_country = df.groupby('Province/State').sum()
confirmed_by_country
```

```python
confirmed_by_country.sample(15)
```

```python
import matplotlib.pyplot as plt
```

### A plotagem leva a decisões erroneas

```python
confirmed_by_country.loc['Beijing'][2:].plot()
plt.title('Casos confirmados Beijing')
plt.figure(figsize=(10,6))
plt.show()
```

```python
plt.figure(figsize=(2,6))
confirmed_by_country.loc['Beijing'][2:].plot()
plt.title('Casos confirmados Beijing')
plt.show()
```

* Nesse caso, o segundo gráfico dá uma intuição de uma alta muito forte
