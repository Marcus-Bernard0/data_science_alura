```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (12, 8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1) #posição 1
g2 = area.add_subplot(2, 2, 2) #posição 2
g3 = area.add_subplot(2, 2, 3) 
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor X Área')

g2.hist(dados['Valor'])
g2.set_title('Histograma')

dados_g3 = dados.Valor.sample(100) #cria uma amostra de 100
dados_g3.index = range(dados_g3.shape[0]) #Refazendo o index
g3.plot(dados_g3)
g3.set_title("Amostra")

grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor médio por tipo')
```
